# 5.Seperate Dependent and independent variables

```python
y = df.iloc[:,1:2]
```

```python
df = df.drop('Attrition',axis = 1)
x = df.iloc[:,:]
```

```python
y.head()
```

```python
x.head()
```

# 6.Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for column in x.columns:
    if x[column].dtypes == "object":
        x[column] = le.fit_transform(x[column])
```

```python
x.head()
```

```python
for column in y.columns:
    if y[column].dtypes == "object":
        y[column] = le.fit_transform(y[column])
```

```python
y.head()
```

# 7.splitting into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# 8.Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
x_train = pd.DataFrame(ms.fit_transform(x_train),columns=x_train.columns)
x_test = pd.DataFrame(ms.fit_transform(x_test),columns=x_test.columns)
y_train = pd.DataFrame(ms.fit_transform(y_train),columns=y_train.columns)
y_test = pd.DataFrame(ms.fit_transform(y_test),columns=y_test.columns)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# 9.Import the model building Libraries

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(max_depth=2, random_state=0)
```

# 10.Initializing the model

```python
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
pred1
```

```python
model2.fit(x_train,y_train)
pred2=model2.predict(x_test)
pred2
```

```python
model3.fit(x_train,y_train)
pred3=model3.predict(x_test)
pred3
```

# 11.Evaluation of Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
print(accuracy_score(y_test,pred1))
print(confusion_matrix(y_test,pred1))
print(classification_report(y_test,pred1))
probability1=model1.predict_proba(x_test)[:,1]
print(probability1)
fpr,tpr,threshsholds = roc_curve(y_test,probability1)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

```python
print(accuracy_score(y_test,pred2))
print(confusion_matrix(y_test,pred2))
print(classification_report(y_test,pred2))
probability2=model2.predict_proba(x_test)[:,1]
print(probability2)
fpr,tpr,threshsholds = roc_curve(y_test,probability2)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

```python
print(accuracy_score(y_test,pred3))
print(confusion_matrix(y_test,pred3))
print(classification_report(y_test,pred3))
probability3=model3.predict_proba(x_test)[:,1]
print(probability3)
fpr,tpr,threshsholds = roc_curve(y_test,probability3)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
